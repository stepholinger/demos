```python
import obspy
from obspy.clients.fdsn import Client
from obspy.signal.cross_correlation import correlate_stream_template, correlation_detector
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
```

```python
'''

Download some seismic data that contains a Whillans slip event from Pratt et al 2014

'''

# choose time window of interest
t = obspy.UTCDateTime("2010-12-16T13:20:00.000")
dt = 3600*2

# make query to FDSN (International Federation of Digital Seismograph Networks)
client = Client("IRIS")
net = "YT"
stat = "SURP"
chan = "BH*"

# pull data and remove response
st_event = client.get_waveforms(network=net, station=stat, location="*", channel=chan, starttime=t, endtime=t+dt,
                    attach_response=True)

# remove instrument response to displacement
st_event.remove_response(output="DISP")

# basic preprocessing
st_event.detrend("demean")
st_event.detrend("linear")
st_event.taper(max_percentage=0.1, max_length=200.)

# bandpass filter 
freq = [1/100,1/30]
st_event.filter("bandpass",freqmin=freq[0],freqmax=freq[1])

# plot the data
plot_start = obspy.UTCDateTime("2010-12-16T13:30:00.000")
plot_end = plot_start + 3600/2
st_event.plot(starttime=plot_start,endtime=plot_end);
```

```python
'''

Download a longer snippet of data that contains multiple events

'''

# choose time window of interest
t = obspy.UTCDateTime("2010-12-10T00:00:00.000")
dt = 86400*4

# pull data and remove response
st_long = client.get_waveforms(network=net, station=stat, location="*", channel=chan, starttime=t, endtime=t+dt,
                    attach_response=True)

# remove instrument response to displacement
st_long.remove_response(output="DISP")

# basic preprocessing
st_long.detrend("demean")
st_long.detrend("linear")
st_long.taper(max_percentage=0.1, max_length=200.)

# bandpass filter 
freq = [1/100,1/30]
st_long.filter("bandpass",freqmin=freq[0],freqmax=freq[1])

# plot the data
st_long.plot();
```

```python
'''

Choose a template waveform for use in detection

'''

# select desired station and channel for template
stat_temp = "SURP"
chan_temp = "BHN"
st_temp = st_event.select(station=stat_temp,channel=chan_temp)

# select time window for template
temp_start = obspy.UTCDateTime("2010-12-16T13:30:00.000")
temp_end = obspy.UTCDateTime("2010-12-16T14:00:00.000")
st_temp = st_temp.slice(starttime=temp_start,endtime=temp_end)
```

```python
'''

Plot the template

'''

# get time values and underlying data for plotting (template)
data_temp = st_temp.select(station=stat_temp,channel=chan_temp)[0].data
t_temp = st_temp[0].times(type="matplotlib")

# get time values and underlying data for plotting (longer snippet)
data_event = st_event.select(station=stat_temp,channel=chan_temp)[0].data
t_event = st_event.select(station=stat_temp,channel=chan_temp)[0].times(type="matplotlib")

# plot the template
fig,ax = plt.subplots()
ax.plot(t_event,data_event,c='k')
ax.plot(t_temp,data_temp,c='r')
ax.set_ylabel("Displacement (m)")
ax.set_xlabel("Time")
ax.set_xlim([t_event[0],t_event[-1]])
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
plt.title("Template waveform ("+stat_temp+" "+chan_temp+")")
plt.show()
```

```python
'''

Perform template matching detections on continuous data

'''

# cross correlate template with continuous data
corr = correlate_stream_template(st_long, st_temp)
corr_data = corr[0].data

# set a detection threshold
thresh = 0.65

# plot detection results
fig,ax = plt.subplots()
plot_t = corr[0].times("matplotlib")
plt.plot(plot_t,corr_data,c='k')
plt.hlines(thresh,np.min(plot_t),np.max(plot_t),colors='r',linestyle='--')
ax.set_ylabel("Correlation")
ax.set_xlabel("Time")
ax.set_xlim(np.min(plot_t),np.max(plot_t))
ax.set_ylim(0,1)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d\n%H:%M'))
```

```python
'''

Retrieve and plot the detected events

'''

# get times where correlation function exceeds threshold
thresh_ind = [(corr_data > thresh) == True][0]
thresh_times = corr[0].times()[thresh_ind]
thresh_times = np.array([corr[0].stats.starttime+sec for sec in thresh_times])

# find first time for each region of the correlation function that exceeds the threshold
event_times = []
event_times.append(thresh_times[0])
dt = 1/corr[0].stats.sampling_rate
event_times.extend(thresh_times[1:][(np.diff(thresh_times) > 0.025)])

# retrieve event waveforms
duration = temp_end-temp_start
event_waveforms = []
for event_time in event_times:
    event_waveforms.append(st_long.slice(starttime=event_time,endtime=event_time+duration))

# plot event waveforms
for event_waveform in event_waveforms:
    event_waveform.plot();
```
